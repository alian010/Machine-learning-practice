```python
reg.fit(df[['area']], df[['rent']])
```

```python
reg.predict([[2050]])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df.area, df.rent, color='black', marker='o')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Veg Price')
```

```python
df2 = df.copy()
df2.head()
```

```python
df2['predicted_rent'] = reg.predict(df[['area']])
df2.head()
```

```python
df2['res'] = df2['rent'] - df2['predicted_rent'] #residuals
df2.head()
```

```python
reg.score(df[['area']], df[['rent']]) #r squared value
```
